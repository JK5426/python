```python
np.linspace(2,20) # here start and stop value is maneratry and if num is not pass then it concidered 50 by defualt
```

```python
import numpy as np
np.linspace(2,18,num=10) # it divide the array item in 10 interval and it returns float type by defualt 
```

```python
# int type return
np.linspace(2,20,num=10,dtype="int")
```

```python
#see also

a=np.linspace(2,20,num=5,dtype="int")
b=np.linspace(2,20,num=5)
print(a)# return int type bcoz it is mensioned
print(b)# return float type by defualt
```

```python
np.linspace(2,20,num=10,endpoint=False) # if endpoint is False then then not concidered the endpoint by default it is True
```

```python
np.linspace(2,20,num=10,endpoint=True) # it concidered the last point
```
